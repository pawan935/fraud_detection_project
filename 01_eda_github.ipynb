```python
#Import libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#Load dataset from data folder
df = pd.read_csv("../data/creditcard.csv", encoding="latin1")
```

```python
#overview
df.shape
df.head()
```

```python
#statistics
df.info()
df.describe()
```

```python
#Transaction Amount overview
plt.Figure(figsize=(10,5))
sns.histplot(df['Amount'],bins=50,kde=True)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Count")
plt.show
```

```python
#Transaction Amount by Class
plt.Figure(figsize=(10,5))
sns.histplot(data=df,x='Amount',bins=50,hue='Class',log_scale=True,kde=True)
plt.title("Transaction Amount by Class (Fraud vs Non-Fraud)")
plt.xlabel("Amount")
plt.ylabel("Count (log scale)")
plt.show
```

```python
#Class balance visualization
plt.figure(figsize=(6,4))
sns.countplot(x='Class',data=df)
plt.title("Class Distribution (0 = Non-Fraud, 1 = Fraud)")
plt.show
```

```python
df['Hour'] = (df['Time']//3600)%24  # modulo 24 to get hour of the day
```

```python
# Plot overall transaction counts by hour
plt.figure(figsize=(12,5))
sns.histplot(df['Hour'],bins=24,kde=True)
plt.title("Transactions per Hour (All)")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Transactions")
plt.show
```

```python
# Plot transactions by hour separated by Class
plt.Figure(figsize=(12,5))
sns.histplot(data=df,x='Hour',bins=24,hue='Class',multiple='stack')
plt.title("ransactions per Hour by Class (Fraud vs Non-Fraud)")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Transactions")
plt.show
```

```python
#Fraud ratio per hour
hourly_fraud_ratio =df.groupby('Hour')['Class'].mean()
plt.Figure(figsize=(12,5))
hourly_fraud_ratio.plot(kind='bar')
plt.title("Fraud Ratio per Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Fraud Ratio (Fraction of Transactions")
plt.show
```

```python
#Select numeric columns only
numeric_df = df.select_dtypes(include=['float64','int64'])
#Compute correlation matrix
corr = numeric_df.corr()
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(corr, cmap="coolwarm_r", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap (All Features)")
plt.show()
```

```python
corr_claass = corr['Class'].sort_values(ascending=False)
print("Top features positively correlated with Class (Fraud):")
print(corr_claass.head(10))
print("\nTop features negatively correlated with Class (Fraud):")
print(corr_claass.tail(10))
```

```python
# Compute correlation with Class
corr_class = numeric_df.corr()['Class'].sort_values(ascending=False)
top_features = corr_class.index[1:6]  # top 5 features excluding 'Class'

print("Top 5 features correlated with Fraud:", list(top_features))
```

```python
for feature in top_features:
    plt.figure(figsize=(8,4))
    sns.boxplot(x='Class', y=feature, data=df)
    plt.title(f"{feature} distribution by Class (0 = Non-Fraud, 1 = Fraud)")
    plt.show()
```

```python
for feature in top_features:
    plt.figure(figsize=(8,4))
    sns.scatterplot(x=feature, y='Amount', hue='Class', data=df, alpha=0.6)
    plt.title(f"{feature} vs Amount (Fraud vs Non-Fraud)")
    plt.show()
```
